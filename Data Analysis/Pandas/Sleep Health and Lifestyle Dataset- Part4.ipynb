```python
import pandas as pd
```

```python
#Sort

# Data Source Link: https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
#DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind=’quicksort’, na_position=’last’)
#by = column names to sort by that
#axis  = 0 rows, 1 for columns
#if inplace = True, will make changes in passed data
#kind: sort algorithm, include quicksort, mergesort or heapsort
#na_position: 'last' or 'first' to set position of null values. 'last' is default 

#first let's see the data
data
```

```python
#Now let's sort using the function. I want to sort  by Heart Rate

out = data.sort_values("Heart Rate", axis = 0, ascending = True, inplace = False, na_position= 'last')

#let's see the results
out
```

```python
#Now let's sort different!
#We want to sort different columns, but each in order we want (ascending or descending)
# Sort two columns, Physical Activity Level and Heart Rate, first ascending next descending 
out2 = data.sort_values(by= ["Physical Activity Level", "Heart Rate"], axis = 0,
                        ascending=[True, False], inplace= False)

out2
```
