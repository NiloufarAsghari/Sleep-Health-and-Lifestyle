```python
import pandas as pd
```

```python
#Truncate the data before and after some index (aka rows)
# Data Source Link: https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

print("index", data.index)
```

```python
#let's see rows from 55 to 65

for i in range (55,65):
    rowret = data.loc[i]
    print(rowret)
```

```python
#Source: geeksforgeeks
#Now, we want to truncate all rows before 55 and after 65
#This means that we'll have only datas from 55-65 after truncation
# Use Data.truncate(before=None, after=None, axis=None, copy=True)
#Axis to truncate is index (rows) by default
#copy:Return a copy of the truncated section

output2 = data.truncate(before = 55 , after = 65)
```

```python
print(output2)
```

```python
#let's check before 55 and after 65

row55 = output2.loc[55]
print(row55)

row65 = output2.loc[65]
row65
```
